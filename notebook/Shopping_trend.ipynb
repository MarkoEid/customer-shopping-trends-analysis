# Dataset Description

- Customer ID – Unique identifier for each customer
- Age – Customer age
- Gender – Customer gender (Male/Female/Other)
- Item Purchased – Name of the item bought
- Category – Category of the item (e.g., Clothing, Electronics)
- Purchase Amount (USD) – Amount spent in US dollars
- Location – Customer’s geographic location
- Size – Size of the product (S, M, L, etc.)
- Color – Color of the product
- Season – Season in which the purchase was made
- Review Rating – Product rating given by the customer
- Subscription Status – Whether the customer has a subscription (Yes/No)
- Payment Method – Method used to pay for the purchase
- Shipping Type – Standard / Express, etc.
- Discount Applied – Indicates if a discount was used
- Promo Code Used – Promo code applied (if any)
- Previous Purchases – Number of previous purchases made by the customer
- Preferred Payment Method – The customer’s usual payment method
- Frequency of Purchases – How often the customer makes purchases

# cleaning the data

### Binary Transformation: Converting ‘Yes’ and ‘No’ to 1 and 0 

```python
df["Discount Applied"]
```

```python
df["Subscription Status"] = df["Subscription Status"].map({"Yes": 1, "No": 0})
```

```python
df
```

### Transforming Gender Labels to Short Form

```python
df["Gender"].replace("Male","M",inplace=True)
df["Gender"].replace("Female","F",inplace=True)
```

### Standardizing Frequency Labels

```python
df["Frequency of Purchases"] = df["Frequency of Purchases"].replace({"Fortnightly": "Bi-Weekly"})
df["Frequency of Purchases"] = df["Frequency of Purchases"].replace({"Quarterly": "Every 3 Months"})
```

```python
df["Frequency of Purchases"].unique()
```

###  Renaming Columns to Shorter Names

```python
df.rename(columns={"Item Purchased":"Item"},inplace=True)
df.rename(columns={"Review Rating":"Rating"},inplace=True)
df.rename(columns={"Purchase Amount (USD)":"Price(USD)"},inplace=True)
```
